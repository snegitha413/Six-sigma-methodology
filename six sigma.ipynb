```python
from sklearn.preprocessing import MinMaxScaler
```

```python
trans = MinMaxScaler()
```

```python
trans_data1=trans.fit_transform(data1)
trans_data2=trans.fit_transform(data2)
```

```python
from pandas import DataFrame
```

```python
td1=DataFrame(trans_data1)
td2=DataFrame(trans_data2)
```

```python
from scipy import stats
import numpy as np
```

```python
z1 = np.abs(stats.zscore(td1))
```

```python
z2 = np.abs(stats.zscore(td2))
```

```python
print(np.where(z1 > 3))
```

```python
print(np.where(z2 > 3))
```

```python
newd1 = td1[(z1 < 3).all(axis=1)]
```

```python
newd2 = td2[(z2 < 3).all(axis=1)]
```

```python
plt.figure(figsize =(20,20))

plt.subplot(5,2,1)
stats.probplot(newd1[3], plot=plt)
plt.title("residual sugar of red wine")

plt.subplot(5,2,2)
stats.probplot(newd2[3], plot=plt)
plt.title("residual sugar of white wine")

plt.subplot(5,2,5)
stats.probplot(data1['chlorides'], plot=plt)
plt.title("chlorides of red wine")
plt.subplot(5,2,6)
stats.probplot(data2['chlorides'], plot=plt)
plt.title("chlorides of white wine")

plt.show()
```

```python
target = 7
LSL = 3
USL = 9

# Generate normally distributed data points
data = np.random.normal(loc=target,scale=1,size=100)

# Generate probability density function 
x = np.linspace(min(newd1), max(newd1), 1000)
y = norm.pdf(x, loc=5, scale=1)

# Plot histogram for data along with probability density functions and specification limits
plt.figure(figsize=(15,10))
plt.hist(data, color="lightgrey", edgecolor="black", density=True)
sns.kdeplot(data, color="blue", label="Density ST")
plt.plot(x, y, linestyle="--", color="black", label="Theorethical Density ST")
plt.axvline(LSL, linestyle="--", color="red", label="LSL")
plt.axvline(USL, linestyle="--", color="orange", label="USL")
plt.axvline(target, linestyle="--", color="green", label="Target")
plt.title('Process Capability Analysis for red wine')
plt.xlabel("Measure")
plt.ylabel("")
plt.yticks([])
plt.legend()
plt.show()
```
